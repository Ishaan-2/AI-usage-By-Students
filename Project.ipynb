```python
def ai_tools(para):
        if ', ' in para:
            para = para.replace(', ',',')  
            return para
        else:
            return para
    
df['AI_Tools_Used'] = df['AI_Tools_Used'].apply(ai_tools)
ai_counter = Counter()
for tools in df['AI_Tools_Used']:
    ai_counter.update(tools.split(','))
ai_tool=[]
tools_usage=[]
for k,v in ai_counter.items():
    ai_tool.append(k)
    tools_usage.append(v)

print(ai_tool)
tools_usage
```

```python
def use_case(para):
        if ', ' in para:
            para = para.replace(', ',',')  
            return para
        else:
            return para
    
df['Use_Cases'] = df['Use_Cases'].apply(use_case)
use_counter=Counter()
for c in df['Use_Cases']:
    use_counter.update(c.split(','))
usage=[]
n_usage=[]
for k,v in use_counter.items():
    usage.append(k)
    n_usage.append(v)
```
